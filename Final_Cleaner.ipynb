```python
import pandas as pd
import csv
import numpy as np
import re
from collections import Counter
import os
directory = os.fsencode('Presidents_condensed')
```

```python
presidents = {
    'roosevelt': 'Republican',
    'taft': 'Republican',
    'wilson': 'Democrat',
    'harding': 'Republican',
    'coolidge': 'Republican',
    'hoover': 'Republican',
    'fdroosevelt': 'Democrat',
    'truman': 'Democrat',
    'eisenhower': 'Republican',
    'kennedy': 'Democrat',
    'johnson': 'Democrat',
    'nixon': 'Republican',
    'ford': 'Republican',
    'carter': 'Democrat',
    'reagan': 'Republican',
    'bush': 'Republican',
    'gwbush': 'Republican',
    'clinton': 'Democrat',
    'obama': 'Democrat'    
}
```

```python
rep = {'.':'', ',':'', ':':'', ';':'', '"':'', "'":'', '?':'', '-':'', '--':'', '/':'', '$':''}

def replace_all(text, dic):
    for i, j in dic.items():
        text = text.replace(i, j)
    return text
```

```python
with open("Presidents_condensed/clinton_speeches_013.txt", encoding="utf8") as file:
    data = file.read() 
    split_speech = data.splitlines(True)[2:]
    final_list = []
    for i in split_speech:
        i = i.strip('\n')
        if i == '':
            pass
        else:
            final_list.append(i)
            
    speech = ' '.join(final_list)
    speech = speech.lower()
    speech = replace_all(speech, rep)
    print(speech)
```

```python
speech1 = []
pres = []
party = []
length = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".txt"):
        with open("Presidents/{}".format(filename), encoding="utf8") as file:
            data = file.read()
            if len(data) < 30000:
                split_speech = data.splitlines(True)[2:]
                final_list = []
                for i in split_speech:
                    i = i.strip('\n')
                    if i == '':
                        pass
                    else:
                        final_list.append(i)

                speech = ' '.join(final_list)
                speech = speech.lower()
                speech = replace_all(speech, rep)

                speech1.append(speech)
                length.append(len(speech))
                pres_name = filename.split('_')[0]
                for key, value in presidents.items():
                    if key == pres_name:
                        pres.append(key)
                        party.append(value)

    else:
        print('oh no')

df = pd.DataFrame({'value': speech1, 'Presidents': pres, 'Party_Affliation': party})
```
