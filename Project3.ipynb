```python
presidents = {
    'roosevelt': 'Republican',
    'taft': 'Republican',
    'wilson': 'Democrat',
    'harding': 'Republican',
    'coolidge': 'Republican',
    'hoover': 'Republican',
    'fdroosevelt': 'Democrat',
    'truman': 'Democrat',
    'eisenhower': 'Republican',
    'kennedy': 'Democrat',
    'johnson': 'Democrat',
    'nixon': 'Republican',
    'ford': 'Republican',
    'carter': 'Democrat',
    'reagan': 'Republican',
    'bush': 'Republican',
    'gwbush': 'Republican',
    'clinton': 'Democrat',
    'obama': 'Democrat'    
}
```

```python
import os
directory = os.fsencode('Presidents')

speech1 = []
speech2 = []
speech3 = []
speech4 = []
speech5 = []
speech6 = []
pres = []
party = []
length = []

for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.endswith(".txt"):
        with open("Presidents/{}".format(filename), encoding="utf8") as file:
            data = file.read() 
            data = data.replace(',', ' ')
            split_speech = data.splitlines(True)[2:]
            speech = ' '.join(split_speech)
            speech = speech.replace('.', ' ')
            speech = speech.replace('>', ' ')
            speech = speech.replace('<', ' ')
            speech = speech.replace(':', ' ')
            speech = speech.replace('the ', ' ')
            speech = speech.replace('Thank you ', ' ')
            speech = speech.replace('thank you ', ' ')
            speech = speech.replace('Speaker ', ' ')
            speech = speech.replace('PRESIDENT ', ' ')
            speech = speech.replace('president ', ' ')
            speech = speech.replace('President ', ' ')
            speech = speech.replace('Good evening ', ' ')
            speech = speech.replace('I ', ' ')
            speech = speech.replace('Mr ', '')
            speech = speech.replace('GENTLEMEN OF THE CONGRESS ', '')
            speech = speech.replace('it ', '')
            #speech = speech.strip('\n')
            #speech = speech.strip()
            speech = speech.replace('\n', ' ')
            
            pattern = re.compile(r'\s+')
            speech = re.sub(pattern, ' ', speech)
            speech = speech.strip()
            
            speech1.append(speech[0:29000])
            
            if speech[29000:58000]:
                speech2.append(speech[29000:58000])
            else:
                speech2.append('l')
                
            if speech[58000:87000]:
                speech3.append(speech[58000:87000])
            else:
                speech3.append('l')
                
            if speech[87000:116000]:
                speech4.append(speech[87000:116000])
            else:
                speech4.append('l')
                
            if speech[116000:145000]:
                speech5.append(speech[116000:145000])
            else:
                speech5.append('l')
                
            if speech[145000:174000]:
                speech6.append(speech[145000:174000])
            else:
                speech6.append('l')
                
            length.append(len(speech))
            pres_name = filename.split('_')[0]
            for key, value in presidents.items():
                if key == pres_name:
                    pres.append(key)
                    party.append(value)
    else:
        print('oh no')
        
df = pd.DataFrame({'Speech1': speech1, 'Speech2': speech2, 'Speech3': speech3,
                   'Speech4': speech4, 'Speech5': speech5, 'Speech6': speech6,
                   'Presidents': pres, 'Party_Affliation': party})
        
```

```python
df = df.melt(['Presidents', 'Party_Affliation'])
```

```python
df = df.loc[df['value']!='l']
```

```python
df.columns = df.columns.str.strip()
```
